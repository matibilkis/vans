```python
energies = {}
histories = {}
resolvers = {}
J=2.0
args={"n_qubits":8,"problem_config":{"problem" : "XXZ", "g":1.0, "J": J}, "load_displaying":False, "specific_folder_name":"8Q - J {} g 1.0".format(J)}
vqe_handler = VQE(n_qubits=args["n_qubits"],problem_config=args["problem_config"],optimizer="adam", lr=1., epochs = 10**4, patience=500)
ground_energy = np.min(np.linalg.eigvals(sum(vqe_handler.observable).matrix()))

for optimizer in ["adam","sgd", "adagrad"]:
    
    energies[optimizer] = {}
    histories[optimizer] = {}
    resolvers[optimizer] = {}
    
    for L in tqdm(range(3,15)):

        vqe_handler = VQE(n_qubits=args["n_qubits"],problem_config=args["problem_config"],optimizer=optimizer, lr=1., epochs = L*10**3, patience=100*L, max_vqe_time=300)
        ground_energy = np.min(np.linalg.eigvals(sum(vqe_handler.observable).matrix()))

        indexed_circuit = vqe_handler.hea_ansatz_indexed_circuit(L=L)
        energy, resolver, history = vqe_handler.vqe(indexed_circuit)
        
        energies[optimizer][L] = energy
        histories[optimizer][L] = resolver
        resolvers[optimizer][L] = history
        
```
