```python
data = np.load("H2/fcisH2.npy")
data1 = np.load("H2/fcisH2_points16.npy")

js=np.linspace(0.5,2.3,16)
energies=[]
iterations=[]
for j in js:
    print("BOND LENGTH: {}".format(j))
    arguments={"n_qubits":4, "bond_length":j}
    evaluator = Evaluator(loading=True, args=arguments)
    energies.append(evaluator.raw_history[len(list(evaluator.raw_history.keys()))-1][-1])
    iterations.append(len(list(evaluator.raw_history.keys())))
    display.display(evaluator.raw_history[len(list(evaluator.raw_history.keys()))-1][0])
    print("\n")
```

```python
plt.figure(figsize=(15,5))
for k in range(3):

    ax = plt.subplot2grid((1,3),(0,k))
    if k==0:
        ax.plot(data[0],data[1],label="FCI")
        ax.scatter(js,energies,marker='*',s=100,color="black",label="Vans - 60 runs")
        ax.set_ylabel("Energy [Ha]",size=20)
    elif k==1:
        ax.plot(js,energies-data1[1])
        ax.scatter(js, energies-data1[1],s=100,label="Energy shift")
        ax.plot(js,np.ones(len(js))*0.0016,'--',color="black",label="Chemical accuracy")
        ax.yaxis.tick_right()
    else:
        for v,b in enumerate(iterations):
            ax.bar(v,b)
        ax.yaxis.tick_right() 
        
    ax.legend()

    ax.set_xlabel("Bond Lenght [Å]",size=20)
plt.savefig("second_run.png")
```
